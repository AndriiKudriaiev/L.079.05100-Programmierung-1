# Aufgabe 1: Kokosnüsse

Drei Seeleute und ein Affe sammeln auf einer Insel Kokosnüsse, die am nächsten Tag geteilt werden sollen. 

In der Nacht teilt der erste Seemann in Sorge um seinen Anteil die Nüsse in drei gleich große Teile und nimmt sich seinen Anteil. Eine Nuss, die bei der Teilung übrig bleibt, bekommt der Affe. 

Kurze Zeit später macht der zweite Seemann das gleiche mit dem Rest. Wieder bleibt eine Nuss für den Affen übrig. 

Auch der dritte Seemann teilt den Rest noch einmal und gibt dem Affen eine übrig gebliebene Nuss. 

Am nächsten Morgen gibt keiner zu, dass er seinen Anteil schon hat, sondern alle teilen die verbliebenen Nüsse noch einmal unter sich auf und geben dem Affen eine restliche Nuss.

## Aufgabenstellung

Schreiben Sie ein Python-Programm mit folgenden Funktionen:

- Der Seemann nimmt 1/3 der Nüsse, der Affe bekommt 1 Nuss
- Dies passiert dreimal in der Nacht, dann einmal am Morgen

Schreiben Sie eine Funktion find_root(lower, upper), die mit dem Intervallhalbierungsverfahren eine Nullstelle der Sinusfunktion im Intervall [lower, upper] sucht.

Annahme: Sie sollen voraussetzen, dass zwischen den Anfangswerten lower und upper genau eine Nullstelle liegt.
Das bedeutet unter anderem, dass sin(lower) und sin(upper) verschiedene Vorzeichen haben.

Das Verfahren soll das Suchintervall solange halbieren, bis die Intervallgrenzen um weniger als epsilon = 0.0001 auseinander liegen.

Die Funktion soll am Ende den Mittelwert des finalen Intervalls zurückgeben, also: (lower + upper) / 2

Testen Sie Ihre Funktion im Hauptprogramm mit folgenden Testfällen:

    Suche nach einer Nullstelle zwischen 3 und 4
    Suche nach einer Nullstelle zwischen -1 und 1

```python
# diese Zelle Ausführen aber nicht verändern
import math
```

```python
#Implementieren Sie hier ihre Funktion
#Your Code here
def find_root(lower, upper):
    epsilon = 0.0001
    while (upper - lower) > epsilon:
        mid = (lower + upper)/2
        if math.sin(lower) * math.sin(mid) > 0:
            lower = mid
        else:
            upper = mid
    return (lower + upper)/2

        
```

```python
# Testen Sie Ihre Funktion mit verschiedenen Intervallen.
for bounds in ((3, 4), (-1, 1)):
    a, b = bounds
    root = find_root(a, b)
    print(f"Intervall [{a}, {b}] -> Nullstelle bei {root:.5f}")
```

# Aufgabe 3: Goldbach-Vermutung

## Hintergrund

Die Goldbach-Vermutung ist eines der ältesten ungelösten Probleme der Mathematik. Sie besagt:

**Jede gerade Zahl größer als 2 lässt sich als Summe zweier Primzahlen darstellen.**

Beispiele:    
4 == 2 + 2   
6 == 3 + 3   
8 == 3 + 5   
10 == 3 + 7   
12 == 5 + 7  

Obwohl diese Vermutung für sehr große Zahlen überprüft wurde, gibt es bis heute keinen mathematischen Beweis.

## Aufgabenstellung

Schreiben Sie ein Programm, das die Goldbach-Vermutung für gerade Zahlen überprüft.

Implementieren Sie folgende Funktionen:

1. `is_prime(n)` - Prüft, ob eine Zahl n eine Primzahl ist
2. `find_goldbach_pair(n)` -
    Findet zwei Primzahlen, die sich zu n addieren, und gibt die kleinere der beiden Primzahlen zurück
    oder gibt None zurück, falls keine 2 Primzahlen gefunden werden
4. `verify_goldbach(start, end)` - Überprüft die Goldbach-Vermutung für alle geraden Zahlen n von start bis end
    und gibt n als Summe der beiden Primzahlen aus, in dem Format wie in den Beispielen oben

```python
# Implementieren Sie hier `is_prime`

def is_prime(n):
    if n.is_integer():
        if n==2:
            return True
        elif n>2:
            i=2
            while i!=n:
                if (n/i).is_integer():
                    return False
                i+=1
            return True
        else:
            return False
            
    # YOUR CODE HERE
    raise NotImplementedError()
```

```python
assert is_prime(2)
assert is_prime(19)
assert not is_prime(1)
assert not is_prime(21)
```

```python
# Implementieren Sie hier `find_goldbach_pair`

def find_goldbach_pair(even_number):
    if int(even_number) >= 2:
        a=2
        b=int(even_number)-2
        for i in range(even_number):
            if is_prime(a) is True and is_prime(b):
                return a
            a+=1
            b-=1
        return None
            
    raise NotImplementedError()
```

```python
assert find_goldbach_pair(4) == 2
assert find_goldbach_pair(10) == 3
```

```python
# Implementieren Sie hier `verify_goldbach`

def verify_goldbach(start, end):
    if start%2==0 and end%2==0:
        list_of_values=[]
        start_value = start
        while start_value <= end:
            a=find_goldbach_pair(start_value)
            b=end-a
            print(a,b)
            if is_prime(b):
                list_of_values.append(f"{start_value} = {a} + {b}")
            start_value+=2
        return list_of_values
    
```
