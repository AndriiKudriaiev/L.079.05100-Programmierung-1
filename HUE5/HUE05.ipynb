## Aufgabe 1: Rotation nach links

Schreiben Sie eine Funktion `rot`, die ein Tupel (oder einen String) als Parameter bekommt, und ein Tupel (oder einen String) zurückgibt, in der die Elemente (oder die Zeichen) um eine Position nach links rotiert sind, d.h., das erste Element (oder das erste Zeichen) kommt ans Ende des Tupels (oder des Strings).   
Beispiele: 

Beispielsweise soll der Aufruf
```python
rot("abcd")
``` 
sollte den Wert `'bcda'` zurückgeben und
```python
rot(((1,2),(3,4),5))
```
den Wert  `((3, 4), 5, (1, 2))`

````python
# Implementieren Sie hier Ihre Funktion

def rot(s):
    if len(s) <= 1:
        return s
    return s[1:]+s[:1]
````

````python
# Tests für `rot`
assert rot("abcd") == "bcda", 'public test failed: rot("abcd") == "bcda"'
assert rot(((1, 2), (3, 4), 5)) == ((3, 4), 5, (1, 2)), "rot should also work for tuples"
````

## Aufgabe 3: Flächeninhalt von Polygonen 

Schreiben Sie eine Funktion `flächeninhalt`, welche als Parameter ein Polygon in Form eines Tupels an Punkten $(x,y)$ übergeben bekommt. Die Funktion soll den Flächeninhalt des übergebenen Polygons berechnen und zurückgeben. Ein Rechteck mit Breite 1 und Höhe 2, dessen untere linke Ecke im Punkt $(0,0)$ liegt, könnte zum Beispiel so wie in Zelle 1 gegeben sein. Sie können davon ausgehen, dass die Kanten kreuzungsfrei sind. 

Wenn
$$
(x_0,y_0) , (x_1,y_1), \dots, (x_n,y_n), (x_0,y_0)
$$
die Eckpunkte des Polygons sind (der $n+1$-te Punkt ist gleich 
dem $0$-ten), dann lautet die Formel für die Fläche: 
$$
    \frac{x_0 y_1 - y_0 x_1}{2} + \frac{x_1 y_2 - y_1 x_2}{2} + \frac{x_2 y_3 - y_2 x_3}{2} + \dots + \frac{x_n y_0 - x_0 y_n}{2}
    %% (X0*Y1 - Y0*X1)/2 + (X1*Y2 - Y1*X2)/2 + (X2*Y3 - Y2*X3)/2 + ... + (Xn*Y0 - X0*Yn)/2 
$$

Beispielsweise soll der Aufruf

```python
flächeninhalt(((0, 0), (1, 0), (1, 2), (0, 2)))
```
den Wert `2.0` zurückgeben.

````python
# Implementieren Sie hier Ihre Funktion

def flächeninhalt(polygon):
    """Berechnet den Flächeninhalt eines Polygons mittels Shoelace-Formel."""
    # YOUR CODE HERE
    fläche=0
    for i in range(len(polygon)):
        output=0
        if i == range(len(polygon)):
            x0,y0=polygon[i]
            x1,y1=polygon[0]
            fläche+=(x0*y1-y0*x1)/2
        else:
            x0,y0=polygon[i-1]
            x1,y1=polygon[i]
            fläche+=(x0*y1-y0*x1)/2
    return fläche
    raise NotImplementedError()
````

## Aufgabe 4: Duplikate in einer Liste von rechts entfernen (ohne vordefinierte Funktion zu nutzen) 
Programmieren Sie eine Funktion `remove_duplicates`, die eine Liste `L` als Parameter übergeben bekommt und, von rechts ausgehend, nach Duplikaten in der Liste sucht und alle Duplikate in der Liste löscht.    
Gemeint ist, dass bei Duplikaten das am weitesten rechts in der Liste stehende Element erhalten bleiben soll 
und die weiter links stehenden gleichen Elemente gelöscht werden sollen.    

Des Weiteren sollen Sie KEINE vordefinierten Funktionen verwenden, um diese Aufgabe zu lösen.

**Wichtig**: Es soll keine neue Liste ausgegeben werden, sondern die als Parameter übergebene Liste soll geändert werden. 

Beispielsweise soll nach dem Aufruf
```python
L = [4, 1, 4, 2, 10, 8, 2, 4, 5, 6, 8]
remove_duplicates(L)
```
gelten, dass
```python
L == [1, 10, 2, 4, 5, 6, 8]
```

````python
# Implementieren Sie hier Ihre Funktion

def remove_duplicates(L):
    # YOUR CODE HERE
    i=len(L)-1
    while i>0:
        j=i-1
        while j>=0:
            if L[j] == L[i]:
                del L[j]
                i-=1
            j-=1
        i-=1
    return L
````

````python
# Tests for `remove_duplicates`
data = [4, 1, 4, 2, 10, 8, 2, 4, 5, 6, 8]
remove_duplicates(data)
assert data == [1, 10, 2, 4, 5, 6, 8], "List should contain only rightmost duplicates"
````
